```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url1 = "https://raw.githubusercontent.com/HarshTiwari1710/Sobershot/refs/heads/main/all_drinks.csv"
url2 = "https://raw.githubusercontent.com/HarshTiwari1710/Sobershot/refs/heads/main/mr-boston-flattened.csv"
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
```

```python
df2 = df2.rename(columns={
    'name': 'strDrink',
    'category': 'strCategory',
    'ingredient-1': 'strIngredient1', 'measurement-1': 'strMeasure1',
    'ingredient-2': 'strIngredient2', 'measurement-2': 'strMeasure2',
    'ingredient-3': 'strIngredient3', 'measurement-3': 'strMeasure3',
    'ingredient-4': 'strIngredient4', 'measurement-4': 'strMeasure4',
    'ingredient-5': 'strIngredient5', 'measurement-5': 'strMeasure5',
    'ingredient-6': 'strIngredient6', 'measurement-6': 'strMeasure6',
    'instructions': 'strInstructions',
    'glass': 'strGlass'
})
```

```python
df2
```

```python
missing_columns = [col for col in df1.columns if col not in df2.columns]
for col in missing_columns:
    df2[col] = None
```

```python
df2 = df2[df1.columns]
```

```python
combined_df = pd.concat([df1, df2], ignore_index=True)
```
